```python
import base64
import requests

# OpenAI API Key

#----------------------------------------------
def caption_with_chatg(txt_prompt, target_text, api_key):

    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {api_key}"
    }

    payload = {
        "model": "gpt-4o",
        "messages": [
        {
            "role": "user",
            "content": [
            {
                "type": "text",
                "text": f"{txt_prompt}"
            },
            {
                "type": "text",
                "text": f"{target_text}"
            },
           
            ]
        }
        ],
        "max_tokens": 300
    }

    response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload).json()
    #
    z = response['choices'][0]['message']['content']
    print("content :",z)
    return z 
```
